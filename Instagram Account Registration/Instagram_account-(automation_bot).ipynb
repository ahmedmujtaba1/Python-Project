```python
def login_gmail(driver, wait, email, passw, rec_email_): 
    time.sleep(16)
    try:
        wait.until(EC.presence_of_element_located((By.XPATH, f"//input[@type='email']"))).send_keys(email)
        time.sleep(3)
        wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Next']/.."))).click()
        time.sleep(12)
    except:
        try:
            wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Next']/.."))).click()
        except:pass
                    
    time.sleep(15)
    wait.until(EC.presence_of_element_located((By.XPATH, f"//input[@type='password']"))).send_keys(passw)
    
    wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Next']/.."))).click()

    time.sleep(2)
    
    try:
        driver.find_element(By.XPATH, "//*[text()='Wrong password. Try again or click Forgot password to reset it.']")
        print('The Password is Wrong for this gmail account')
        return
    except:pass
    
    try:
        driver.find_element(By.XPATH, "//span[text()='Verify that it’s you']")
        print('Asking for Number Verification')
        return
    except:pass
    
    try:
        driver.find_element(By.XPATH, "//*[text()='Recover account']")
        print('Account Is Not Verified')
        return
    except:pass
    
    time.sleep(10)
    try:
        wait.until(EC.presence_of_element_located((By.XPATH, f"//div[text()='Confirm your recovery email']/.."))).click()

        time.sleep(6)

        wait.until(EC.presence_of_element_located((By.XPATH, f"//input[@type='email']"))).send_keys(rec_email_)

        time.sleep(2)

        wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Next']/.."))).click()

        time.sleep(2)

        wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Not now']/.."))).click()

        time.sleep(2)
    except:
        try:
            wait.until(EC.presence_of_element_located((By.XPATH, f"//span[text()='Not now']/.."))).click()
        except:pass
```
