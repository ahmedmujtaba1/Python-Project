```python
# Imports
import names
import random
import undetected_chromedriver as uc 
import time
import string
import csv
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
```

```python
chrome_options = uc.ChromeOptions()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument('--disable-popup-blocking')

driver = uc.Chrome(options=chrome_options)
driver.get("https://ichgcp.net/pharma-list/country/greece")
```

```python
with open('Sample1.csv', 'w', newline='',encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(['Company Name','Description','Website','Type of Company','Founded','Headquarters','Area','Products','Revenue ','Number of Employes','FAX','Phone'])
```

```python
actions = ActionChains(driver)
total_count = 0
pages = -1
cnt = 0
num = 1
driver.implicitly_wait(10)

while True:
    try:    
        while pages != 3:
            pages += 1    
            
            scraper = driver.find_elements(By.XPATH,"//span[@class='read_more']")
            
            if cnt == 0:
                last_job = scraper[cnt]
            else:
                last_job = scraper[cnt-1]

            job = scraper[cnt]
            try:
                driver.find_element(By.XPATH,"//div[@id='catfishWrap']//img").click()
                time.sleep(5)
            except:pass

            actions.move_to_element(job).click().perform()
            print('Scraping Details.....')
            
            time.sleep(10)
            try:
                iframe = driver.find_element(By.XPATH,"//iframe[@name='ad_iframe']")
                driver.switch_to.frame(iframe)
                driver.find_element(By.XPATH,"//div[@aria-label='Close ad']").click()
                driver.switch_to.default_content()
            except:pass
                
            time.sleep(5)
            
            try:
                title = driver.find_element(By.XPATH,"//h1").text
                print('Title : ', title)
            
            except:
                title = ''
                
            try:
                nex = driver.find_element(By.XPATH,"//div[@aria-label='Close ad']")
                nex.click()
                print("Closing Add...")
            except:
                try:
                    nexq = driver.find_element(By.XPATH,"//div[@id='dismiss-button']")
                    nexq.click()
                    print("Closing Adds...")
                except:pass
                
            try:
                website = driver.find_element(By.XPATH,"//a[@rel='nofollow']").get_attribute('href')
                print('Website : ', website)
            
            except:
                website = ''
                
            try:
                type_ = driver.find_element(By.XPATH,"((//tr))[1]//td").text
                print('Type : ', type_)
                
            except:
                type_ = ''
                
            try:
                founded = driver.find_element(By.XPATH,"((//tr))[4]//td").text
                print('Founded : ', founded)
                
            except:
                founded = ''
                
            try:
                Headquarters = driver.find_element(By.XPATH,"((//tr))[5]//td").text
                print('Headquarters : ', Headquarters)
                
            except:
                Headquarters = ''
                
            try:
                area = "World Wide"
                print('Area : ', area)
                
            except:
                area = ''
            
            try:
                d = driver.find_element(By.XPATH,"((//tr))[8]//td//ul//li")
                products = []
                for i in range(len(d)):
                    product1 = d[i].text
                    products.append(product1)
                    print('Products : ', products)
            except:
                products = []
                
            try:
                Revenue = driver.find_element(By.XPATH,"((//tr))[9]//td").text
                print('Revenue : ', Revenue)
            
            except:
                Revenue = ''
            
            try:
                employes = driver.find_element(By.XPATH,"((//tr))[14]//td").text
                print('Number of Employes : ', employes)
                
            except:
                employes = ''
            
            driver.back()
            
            time.sleep(15)
            actions = ActionChains(driver)
            name = driver.find_element(By.XPATH,f'(//span[@class="title"]//a)[{num}]')
            try:
                driver.find_element(By.XPATH,"//div[@id='catfishWrap']//img").click()
            except:pass
            actions.move_to_element(name).click().perform()
            time.sleep(5)
            
            try:
                iframe = driver.find_element(By.XPATH,"//iframe[@name='ad_iframe']")
                driver.switch_to.frame(iframe)
                driver.find_element(By.XPATH,"//div[@aria-label='Close ad']").click()
                driver.switch_to.default_content()
            except:pass
            
            time.sleep(5)
            fax = ''
            phone = ''
            
            try:
                b = driver.find_element(By.XPATH,"(//li//p)[2]//b").text
                
                if 'Fax:' in b:
                    fax = driver.find_element(By.XPATH,"(//li//p)[2]//a").text
                    print('FAX : ', fax)
                
                if 'Phone:' in b:
                    phone = driver.find_element(By.XPATH,"(//li//p)[2]//a").text
                    print('Phone Number : ', phone)
            except:pass
                
                
            try:
                a = driver.find_element(By.XPATH,"(//li//p)[3]//b").text
                
                if 'Fax:' in a:
                    fax = driver.find_element(By.XPATH,"(//li//p)[3]//a").text
                    print('FAX : ', fax)
                
                if 'Phone:' in a:
                    phone = driver.find_element(By.XPATH,"(//li//p)[3]//a").text
                    print('Phone Number : ', phone)
            except:pass
            description = ''
            
            try:
                c = driver.find_element(By.XPATH,"(//div[@class='columns container']//h2)[2]").text
                
                d = f'About {title}'
                if d in c:
                    description = driver.find_element(By.XPATH,"(//div[@class='columns container']//p)[5]").text
                    print('Description : ', description)
                
                if d in c:
                    description = driver.find_element(By.XPATH,"(//div[@class='columns container']//p)[6]").text
                    print('Description : ', description)
            except:pass
                
                
            num += 1
            cnt += 1
            last_job = job
            driver.back()
            with open('Sample1.csv', 'a', newline='',encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerow([title, description, website, type_, founded, Headquarters, area, products, Revenue, employes, fax, phone])

            total_count += 1
            print(total_count)
            print('=========================================================================================')
            time.sleep(3)
            
    except:pass
```
