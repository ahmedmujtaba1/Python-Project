# AMAZON SCRAPING

### This below code is for scraping 1 - 50 pages

```python
try:
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.wait import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    from selenium.webdriver.common.action_chains import ActionChains
    from selenium.webdriver.common.keys import Keys
    import undetected_chromedriver as uc
    import csv
    import time
except:
    print("Any Module is not download")
```

```python
chrome_options = uc.ChromeOptions()
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-infobars')

url = 'https://www.amazon.in/s?me=A3AXGEQKABKICM&marketplaceID=A21TJRUUN4KGV'
driver = uc.Chrome(options=chrome_options)
driver.get(url)
```

```python
with open('amazon_scraper.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['Title','Price','MRP','ASIN'])
```

```python
actions = ActionChains(driver)
total_count = 0
pages = -1
cnt = 0
a = driver.find_elements(By.XPATH, "//div[@class='sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col s-widget-spacing-small sg-col-12-of-16']")

driver.implicitly_wait(10)

while True:
    try:    
        while pages != 50:

            scraper = driver.find_elements(By.XPATH, "//*[@class='a-size-medium a-color-base a-text-normal']")

            if cnt == 0:
                last_job = scraper[cnt]
            else:
                last_job = scraper[cnt-1]

            job = scraper[cnt]

            actions.move_to_element(job).click().perform()
            
            try:
                driver.switch_to.window(driver.window_handles[1])
            except:pass
            
            time.sleep(5)
            #time.sleep(1)
            try:
                asin= driver.find_element(By.XPATH, "(//div[@id='averageCustomerReviews'])[1]").get_attribute('data-asin')
            except:
                asin = ' '
            title = driver.find_element(By.XPATH, "//h1[@id='title']").text
            price = driver.find_element(By.XPATH, "(//span[@class='a-price-whole'])[1]").text
            try:
                mrp = driver.find_element(By.XPATH, "(//*[@class='a-price a-text-price'])[3]").text
            except:
                try:
                    mrp = driver.find_element(By.XPATH, "//*[@class='a-price a-text-price']").text
                except:    
                    try:
                        mrp = driver.find_element(By.XPATH, "(//span[@class='a-price a-text-price a-size-base'])[1]").text
                    except:
                        mrp = ' '

            last_job = job

            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            cnt += 1

            print(f'{title}, {price}, {mrp}, {asin}')

            with open('amazon_scraper.csv', 'a', newline='',encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerow([title, price, mrp, asin])

            total_count += 1
            print(total_count)
            print('=========================================================================================')
            time.sleep(3)
            time.sleep(10)

    except:
        try:
            next1 = driver.find_element(By.XPATH, "//*[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
            next1.click()

            cnt = 0 

            time.sleep(7)
            
            print('---------------------------------------------------------')
            pages += 1
            print(f'Page : {pages}')
            print('---------------------------------------------------------')
        except:pass
```
