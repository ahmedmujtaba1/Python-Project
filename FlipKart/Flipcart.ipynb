```python
driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get("https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=samsung+mobiles%7CMobiles&requestId=e4279c9c-4fe7-47e1-a379-dd16121d6b32")
```

```python
driver.maximize_window()
```

```python
actions = ActionChains(driver)

pages = 0   
cnt = 0    
while True:
    try:
        while True:
            
            scraper = driver.find_elements(By.XPATH,"//div[@class='_4rR01T']")
            
            if cnt == 0:
                last_job = scraper[cnt]
            else:
                last_job = scraper[cnt-1]
            
            job = scraper[cnt]

            actions.move_to_element(job).click().perform()
            
            time.sleep(15)
            driver.switch_to.window(driver.window_handles[1])
            
            try:
                title = driver.find_element(By.XPATH,"//h1//span").text
                print("Title : ", title)
            except:
                title = 'None'
                print("Title : ", title)
                
            try:
                actual_price = driver.find_element(By.XPATH,"//div[@class='_30jeq3 _16Jk6d']").text
                print('Actual Price : ', actual_price)
            except:
                actual_price = 'None'
                print('Actual Price : ', actual_price)
            
            try:
                oldprice = driver.find_element(By.XPATH,"//div[@class='_3I9_wc _2p6lqe']").text
                print('Old Price : ', oldprice)

            except:
                oldprice = 'None'
                print('Old Price : ', oldprice)
            
            try:
                perc = driver.find_element(By.XPATH,"//div[@class='_3Ay6Sb _31Dcoz']//span").text
                print('Percentage Discount : ', perc)
            
            except:
                perc = 'None'
                print('Percentage Discount : ', perc)
            
            
            cnt += 1
            
            driver.close()
            
            driver.switch_to.window(driver.window_handles[0])
            
            
            time.sleep(5)
            
            print(cnt)
            print('===========================================================================================')
    except:
        try:
            pagination = driver.find_element(By.XPATH,"//a[@class='_1LKTO3']")
            pagination.click()
            
            time.sleep(10)
            
            cnt = 0
            pages +=1
        
        except:pass
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import csv
from selenium.webdriver.chrome.service import Service
import time
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get("https://itexamanswers.net/it-essentials-ite-v6-0-v7-0-chapter-1-exam-answers-100.html")
```

```python
with open('Web Scraping.csv', 'w', newline='', encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(['Questions','Correct Answer'])
        
import os

# System call
os.system("")

# Class of different styles
class style():
    BLACK = '\033[30m'
    RED = '\033[31m'
    GREEN = '\033[32m'
    YELLOW = '\033[33m'
    BLUE = '\033[34m'
    MAGENTA = '\033[35m'
    CYAN = '\033[36m'
    WHITE = '\033[37m'
    UNDERLINE = '\033[4m'
    RESET = '\033[0m'
```

```python
question_list = driver.find_elements(By.XPATH,"//div[@id='ftwp-postcontent']//p//strong")

cnt = 3
num = 0
no = 0
for i in range(len(question_list)):
    try:
        cnt += 1
        num += 1
        questions = driver.find_element(By.XPATH,f"(//div[@id='ftwp-postcontent']//p//strong)[{cnt}]").text
        print(f'Question {i} : ', '\033[35m', questions, '\033[0m')
        
        
        for n in range(4):
            no += 1
            options = driver.find_element(By.XPATH,f"(//div[@id='ftwp-postcontent']//ul//li)[{no}]").text
            print(f'Options {n} : ', options)
        
        correct = driver.find_element(By.XPATH,f"(//div[@id='ftwp-postcontent']//ul//li//span[@style='color:red'])[{num}]").text
        print('Correct Answer : ', '\033[31m', correct, '\033[0m')
        print("===============================================================================================")
        
        with open('Web Scraping.csv', 'a', newline='', encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow([questions, correct])
            
    except:
        pass
```
