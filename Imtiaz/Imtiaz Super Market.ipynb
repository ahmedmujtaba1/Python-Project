```python
# Imports
import names
import random
import undetected_chromedriver as uc 
import time
import string
import csv
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
global driver

chrome_options = uc.ChromeOptions()
#chrome_options.add_argument('--load-extension=C:/Users/ABC/Downloads/Ahmed Task/proxy_auth_plugin')
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument('--disable-popup-blocking')

driver = uc.Chrome(options=chrome_options)
driver.get("https://shop.imtiaz.com.pk/")
```

```python
with open('ImtiaZ_supermarkeT.csv', 'w', newline='',encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(['Product Name','Stock Quantity','Add to CART','Image URL'])
```

```python
actions = ActionChains(driver)

driver.implicitly_wait(10)
wait = WebDriverWait(driver, 15)

try:
    wait.until(EC.presence_of_element_located((By.XPATH, f"(//button[@title='Open'])[1]"))).click()

    time.sleep(1)

    city = "Karachi"
    city_holder = driver.find_element(By.XPATH,f"//*[text()='{city}']/..")
    actions.move_to_element(city_holder).click().perform()
    
except Exception as ex:
    if ex == 'Message: no such element: Unable to locate element: {"method":"xpath","selector"':
        pass

try:
    city = "Liaquat National Hospital- Gulshan-e-Iqbal"
    try:
        driver.find_element(By.XPATH,"//*[@title='Clear']").click()
    except:pass
    in_ = wait.until(EC.presence_of_element_located((By.XPATH, f"//input[@placeholder='Select Area']")))
    in_.send_keys(city)
    time.sleep(1)

    hanger = driver.find_element(By.XPATH,f"//*[text()='{city}']/..")
    actions.move_to_element(hanger).click().perform()
    
    time.sleep(1)
    
    driver.find_element(By.XPATH,"//*[text()='Select']").click()
    
except Exception as ex:
    pass
    
time.sleep(7)
num = 1
gum = 1
product = 0
cnt = 0
while True:
    try:
        while True:
        
            xpath = driver.find_element(By.XPATH,f"(//div[@class='MuiBox-root mui-style-1bwt8lc'])[{num}]/..")
            actions.move_to_element(xpath).click().perform() 
            print('[+] Going to this : ', driver.find_element(By.XPATH,f"(//div[@class='MuiBox-root mui-style-1bwt8lc'])[{num}]/..").get_attribute('href'))
            time.sleep(5)
            try:
                xpath = driver.find_element(By.XPATH,f"(//div[@class='MuiBox-root mui-style-1bwt8lc'])[{num}]")
                actions.move_to_element(xpath).click().perform()
                num = num + 1
                time.sleep(12)
            except:pass
            
            try:
                scraper = driver.find_elements(By.XPATH,"//div[@class='hazle-product-item_product_item__FSm1N MuiBox-root mui-style-5bkk4b']")
            except:
                try:
                    scraper = driver.find_element(By.XPATH,"//div[@class='hazle-product-item_product_item__FSm1N MuiBox-root mui-style-5bkk4b']")
                except:pass

            print(f'There are {len(scraper)} product(s)')
            try:
                if cnt == 0:
                    last_job = scraper[cnt]

                else:
                    last_job = scraper[cnt]

            except:pass
            job = scraper[cnt]
            cnt += 1
            actions.move_to_element(job).click().perform()

            time.sleep(5)
            print('-=-=-=-=-===-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=')
            title = driver.find_element(By.XPATH,"//h2").text
            print('Title : ', title)

            stock_avaliability = driver.find_element(By.XPATH,"//span[@style='margin-left: 8px; margin-right: 8px;']").text
            print('Stock Avaliablitly : ', stock_avaliability)

            price = driver.find_element(By.XPATH,"//div[@class='HazleProduct_product_item_price_label__WxyGm']//span").text
            print('Price : ', price)
            time.sleep(1)
            driver.find_element(By.XPATH,"//*[text()='Add To Cart']").click()
            time.sleep(3)

            try:
                while True:
                    try:
                        add_to_cart = driver.find_element(By.XPATH,"(//button[@class='MuiButtonBase-root MuiIconButton-root MuiIconButton-sizeMedium HazleProduct_product_item_icon_btn__5kVMn mui-style-4lmkvf'])[2]")
                        add_to_cart.click()
                    except:break
            except:pass

            add_to_cart = driver.find_element(By.XPATH,"//input[@class='MuiInputBase-input MuiOutlinedInput-input MuiInputBase-readOnly mui-style-1x5jdmq']").get_attribute('value')
            print('Add To Cart : ', add_to_cart)

            time.sleep(2)
            img_url = driver.find_element(By.XPATH,f"//img[@alt='{title}']").get_attribute('src')
            print('Image URL : ', img_url)

            print(product)
            print("============================================================================")
            product += 1

            driver.back()

            try:
                scraper = driver.find_element(By.XPATH,"//div[@class='hazle-product-item_product_item__FSm1N MuiBox-root mui-style-5bkk4b']")
                driver.back()
                time.sleep(5)
                gum += 1
                xpath = driver.find_element(By.XPATH,f"(//div[@class='MuiBox-root mui-style-1bwt8lc'])[{gum}]")
                actions.move_to_element(xpath).click().perform()
            except:pass
            
            last_job = job
            
            with open('ImtiaZ_supermarkeT.csv', 'a', newline='',encoding="utf-8") as f:
                writer = csv.writer(f)
                writer.writerow([title, stock_avaliability, add_to_cart, img_url])
    except:
        try:
            scraper = driver.find_element(By.XPATH,"//div[@class='hazle-product-item_product_item__FSm1N MuiBox-root mui-style-5bkk4b']")
            driver.back()
            time.sleep(5)
            gum += 1
            xpath = driver.find_element(By.XPATH,f"(//div[@class='MuiBox-root mui-style-1bwt8lc'])[{gum}]")
            actions.move_to_element(xpath).click().perform()
        except:pass
            
```
