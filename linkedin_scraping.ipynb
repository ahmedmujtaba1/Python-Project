```python
try:
    from time import sleep
    import undetected_chromedriver as uc
    from selenium import webdriver
    from selenium.webdriver.chrome.options import Options
    from selenium.webdriver.common.action_chains import ActionChains
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.common.keys import Keys
    from selenium.webdriver.support import expected_conditions as EC
    from webdriver_manager.chrome import ChromeDriverManager
except:
    print("Any Module is not download")

def Connect_To_Browser():
    url = 'https://www.linkedin.com/jobs/search/?currentJobId=3308450031&geoId=&keywords=Data%20Scientist&location=canada'
    co = webdriver.ChromeOptions()
    co.add_argument("--start-maximized")
    co.add_argument('--disable-gpu')
    co.add_argument('--disable-infobars')
    co.add_experimental_option("detach", True)
    
    driver = driver = webdriver.Chrome(ChromeDriverManager().install(),options=co)
    driver.get(url)
    action = ActionChains(driver)
    
    b = driver.find_elements(By.XPATH, "//*[@data-tracking-control-name='public_jobs_jserp-result_search-card']")
    c = driver.find_elements(By.XPATH, "//*[@class='base-search-card__title']")
    d = driver.find_elements(By.XPATH, "//*[@class='base-search-card__subtitle']")
    e = driver.find_elements(By.XPATH, "//*[@class='job-search-card__location']")
    g = driver.find_elements(By.XPATH, "//*[@class='job-search-card__listdate']")
    
    no_of_pagedowns = 100
    
    elem = driver.find_element(By.XPATH, "//*[@dir='ltr']")

    
    while no_of_pagedowns:
        elem.send_keys(Keys.PAGE_DOWN)
        driver.find_element(By.XPATH, "//*[@aria-label='Load more results']")
        sleep(0.2)
        no_of_pagedowns-=1
    
    jobs_card = driver.find_elements(By.XPATH,"//a[@data-tracking-control-name='public_jobs_jserp-result_search-card']")
    
    
    for scraping in range(len(b)):
        links = b[scraping].get_attribute('href')
        print('Getting links : ', links)
        
        Jobs = c[scraping].text
        print('Jobs You Search : ', Jobs)
        
        Company = d[scraping].text
        print('Company : ', Company)
        
        locations = e[scraping].text
        print('Location Of Company : ', locations)
        
        time = g[scraping].text
        print('Time when this job was post :', time)
        
        for job in jobs_card:
            action.move_to_element(job).click(job).perform()
            try:
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                #clicks on see more
                actions.move_to_element(WebDriverWait(driver,1).until(EC.presence_of_element_located((By.XPATH,"//*[@aria-label='Show more, visually expands previously read content above']")))).click().perform()
            except:
                print('ERROR')
            
            sleep(3)

            details = driver.find_element(By.XPATH,"//*[@class='show-more-less-html__markup']").text
            print('Details : ', details)
            
        print("====================================================================================")
        
Connect_To_Browser()
```
